```python
import numpy as np
from matplotlib import pyplot as plt
import cantera as ct
import sys

module_path = '../model'
if module_path not in sys.path:
    sys.path.append(module_path)
    
from sei_1d_init import SVptr, objs, params, voltage_lookup

model1 = np.loadtxt('Model1.csv',delimiter = ",")
model2 = np.loadtxt('Model2.csv',delimiter = ",")
model3 = np.loadtxt('Model3.csv',delimiter = ",")

t_1 = model1[0,:].T
SV_1 = model1[1:,:].T

t_2 = model2[0,:].T
SV_2 = model2[1:,:].T

t_3 = model3[0,:].T
SV_3 = model3[1:,:].T


sei = objs['SEI']
elyte = objs['elyte']
names = list()
for i in range(1,sei.n_species):
    names.append(sei.species_names[i])
names.append('eps_sei')
names.append('Anode potential')
names.append('SEI potential')
```

```python
ff = 12
lw = 2.0


profiles1 = SV_1[-1,SVptr['Ck sei'][:,1:]]
nspecies = sei.n_species - 1
MolFrac1 = np.zeros_like(profiles1)
for i in np.arange(nspecies):
    MolFrac1[:,i] = profiles1[:,i]/np.sum(profiles1,1)
    
profiles2 = SV_2[-1,SVptr['Ck sei'][:,1:]]
MolFrac2 = np.zeros_like(profiles2)
for i in np.arange(nspecies):
    MolFrac2[:,i] = profiles2[:,i]/np.sum(profiles2,1)    
    
profiles3 = SV_3[-1,SVptr['Ck sei'][:,1:]]
MolFrac3 = np.zeros_like(profiles3)
for i in np.arange(nspecies):
    MolFrac3[:,i] = profiles3[:,i]/np.sum(profiles3,1)    
    
    
font = plt.matplotlib.font_manager.FontProperties(family='Times New Roman',size=ff)
    
fig2 = plt.figure()
ax2 = fig2.add_axes([0.15,0.2,0.75,0.75])
fig2.set_size_inches((3.5,2.25))

colors = ['b','r','g','k']
for i in np.arange(nspecies):
    p1, = ax2.plot(1e9*np.arange(params['Ny'])/params['dyInv'],100*MolFrac1[:,i],'--',color = colors[i])
    p2, = ax2.plot(1e9*np.arange(params['Ny'])/params['dyInv'],100*MolFrac2[:,i],'-.',color = colors[i]) 
    p3, = ax2.plot(1e9*np.arange(params['Ny'])/params['dyInv'],100*MolFrac3[:,i],color = colors[i]) 

#ax2.plot(1e9*np.arange(params['Ny'])/params['dyInv'],100*MolFrac1)
#ax2.plot(1e9*np.arange(params['Ny'])/params['dyInv'],100*MolFrac2, '--')


ax2.legend(['Model 1','Model 2','Model 3'],frameon=False,prop=font,handletextpad=0.15,borderpad=0.05)

font = plt.matplotlib.font_manager.FontProperties(family='Times New Roman',size=ff)

ax2.set_xlim([0,25])
#ax2.set_ylim([0,100])


for tick in ax2.xaxis.get_major_ticks():
    tick.label1.set_fontsize(ff)
    tick.label1.set_fontname('Times New Roman')
for tick in ax2.yaxis.get_major_ticks():
    tick.label1.set_fontsize(ff)
    tick.label1.set_fontname('Times New Roman')

    
ax2.tick_params(axis="y",direction="in",length=5.5,right=True)
ax2.tick_params(axis="x",direction="in",length=5.5,top=True)
ax2.set_ylabel('Mole Fraction (%)',fontsize=ff,family='Times New Roman',labelpad = -1.)
ax2.set_xlabel('SEI Depth (from anode, nm)',fontsize=ff,family='Times New Roman')
plt.savefig('Figure2.pdf',format='pdf',dpi=350)
```

```python
nlayer = 0
#ax1, fig1 = plt.figure()
#ax1 = fig1.add_axes([0.15,0.2,0.75,0.75])
#fig1.set_size_inches((3.5,2.25))
fig1, ax1 = plt.subplots(3, 1, figsize=(3.5, 6.75),sharex=True,sharey=True)


def plot_timeseries(nlayer):
    profiles1 = SV_3[:,SVptr['Ck sei'][nlayer,1:]]
    MolFrac1 = np.zeros((t_1.shape[0],nspecies))
    for ti in np.arange(t_1.shape[0]):
        MolFrac1[ti,:] += profiles1[ti,:]

        MolFrac1[ti,:] = MolFrac1[ti,:]
        MolFrac1[ti,:] = MolFrac1[ti,:]/(1e-24+np.sum(MolFrac1[ti,:]))
    return MolFrac1

MoleFracs = plot_timeseries(0)
colors = ['b','r','g','k']
for i in np.arange(nspecies):
    p1, = ax1[0].plot(t_1,100*MoleFracs[:,i],color=colors[i])

MoleFracs = plot_timeseries(1)
for i in np.arange(nspecies):
    p2, = ax1[1].plot(t_1,100*MoleFracs[:,i],color=colors[i])

MoleFracs = plot_timeseries(2)
for i in np.arange(nspecies):
    p3, = ax1[2].plot(t_1,100*MoleFracs[:,i],color=colors[i])
    
    

for tick in ax1[2].xaxis.get_major_ticks():
    tick.label1.set_fontsize(ff)
    tick.label1.set_fontname('Times New Roman')
for tick in ax1[0].yaxis.get_major_ticks():
    tick.label1.set_fontsize(ff)
    tick.label1.set_fontname('Times New Roman')
for tick in ax1[1].yaxis.get_major_ticks():
    tick.label1.set_fontsize(ff)
    tick.label1.set_fontname('Times New Roman')
for tick in ax1[2].yaxis.get_major_ticks():
    tick.label1.set_fontsize(ff)
    tick.label1.set_fontname('Times New Roman')

for ax in ax1:
    ax.set_ylim((0,100))
    ax.tick_params(axis="y",direction="in",length=5.5,right=True)
    ax.tick_params(axis="x",direction="in",length=5.5,top=False)
    
ax1[0].set_yticks([0, 25, 50, 75,100])
ax1[1].set_yticks([0, 25, 50, 75])
    
ax1[0].tick_params(axis="x",top=True)   
ax1[0].text(33.85,95,'100',fontsize=ff,family='Times New Roman')

fig1.subplots_adjust(hspace=0,left=0.2,bottom=0.1,top=0.975,right=0.95)
plt.setp([a.get_xticklabels() for a in fig1.axes[:-1]], visible=False)
ax1[0].set_xlim((35,45))
ax1[2].set_xlabel('Time (s)',fontsize=ff+4,family='Times New Roman',labelpad = 2.5)
ax1[1].set_ylabel('Mole Fraction (%)',fontsize=ff+4,family='Times New Roman',labelpad = 2.5)
plt.savefig('Figure3_molefrac1.pdf',format='pdf',dpi=350)
```
